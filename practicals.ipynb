Practical 2: Control Structures & Recursion

```python
num = 5
if num % 2 == 0:
    print("Even")
else:
    print("Odd")

for i in range(1, 6):
    print(i)

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```

Practical 5: Pandas Data Cleaning (with fallback dataset)

```python
import pandas as pd

try:
    df = pd.read_csv("data.csv")
except:
    # Synthetic dataset (backup)
    data = {
        "Name": ["A", "B", None],
        "Age": [20, None, 22]
    }
    df = pd.DataFrame(data)

print(df.isnull())

df.fillna(0, inplace=True)
df.drop_duplicates(inplace=True)

print(df.head())
```

Practical 7: Neural Network (MNIST)

```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

(x_train, y_train), (x_test, y_test) = mnist.load_data()

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=1)
model.evaluate(x_test, y_test)
```

Practical 8: K-Means & PCA (with fallback)

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

try:
    df = pd.read_csv("data.csv")
    data = df.select_dtypes(include=['number'])
except:
    data = pd.DataFrame({
        "A": [1,1,1,10,10,10],
        "B": [2,4,0,2,4,0]
    })

kmeans = KMeans(n_clusters=2)
kmeans.fit(data)
print(kmeans.labels_)

pca = PCA(n_components=2)
print(pca.fit_transform(data))
```

Practical 9: Linear Regression (Housing with fallback)

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

try:
    df = pd.read_csv("housing.csv")
    X = df[['area']]
    y = df['price']
except:
    df = pd.DataFrame({
        "area": [1,2,3,4],
        "price": [100,200,300,400]
    })
    X = df[['area']]
    y = df['price']

model = LinearRegression()
model.fit(X, y)

print(model.predict([[5]]))
```
